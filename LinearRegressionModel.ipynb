```python
#Importing necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Formula for univariate linear regression y= wx + b
#Implement linear regression on house price prediction
```

```python
x_train = np.array([1.0,2.0]) #Size in square feet(1000's)
y_train = np.array([300.0,500.0]) #Price in dollars(1000$'s)
```

```python
print(f"Samples in x:{x_train}")
print(f"Samples in y:{y_train}")
```

```python
#Number of training examples
print(f"x_train shape:{x_train.shape}")
m = x_train.shape[0]
print(f"Number of training examples is:{m}")
```

```python
#m is the number of examples
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
#We can observe the data in indices
print(f"x^({x_train[0]}),y^({y_train[0]}) = ({x_train[0]},{y_train[0]})")
print(f"x^({x_train[1]}),y^({y_train[1]}) = ({x_train[1]},{y_train[1]})")
```

```python
#Now we can plot the data
plt.scatter(x_train,y_train, marker="o", color="red")
plt.title("Housing prices")
plt.xlabel('Size (1000 sqft)')
plt.ylabel("Price(1000's) $")
plt.show()
```

```python
#Now we can create linear regression model
def model(x,w,b):
  m = x_train.shape[0]
  f_wb = np.zeros(m)
  for i in range(m):
    f_wb[i] = w * x[i] + b
  return f_wb
```

```python
y_prediction = model(x_train,200,100,)
y_prediction
```

```python
plt.plot(x_train,y_prediction , label="Our Prediction")
plt.scatter(x_train,y_train, marker="o" ,color="red" , label="Actual Values")
plt.title("Housing prices")
plt.xlabel('Size (1000 sqft)')
plt.ylabel("Price(1000's) $")
plt.legend()
plt.show()
```
